```python
# Carry out correlation analysis on the dataset and find the correlation factor
# between
# a. ph and Chloramines
# b. ph and Conductivity
# What do we infer from this correlation value?
```

```python
corrmatrix=data.corr()
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(corrmatrix,vmin=-1,vmax=1,annot=True,linewidth=0.2)
```

```python
## ph and Chloramines - correlation  factor between them is -0.025 ,
## i.e negative correlation-As the value of chloramines increases ph decreases.
```

```python
## The correlation factor between ph and conductivity is 0.014,I.e positive correlation.
## As conductivity increases ph also increases
```

```python
## Since both correlation factors are nearly equal to zero,
## we can say that both quantities has no strong correlation with ph value.
```
